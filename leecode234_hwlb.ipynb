## leecode 234 **[回文链表](https://leetcode.cn/problems/palindrome-linked-list/description/?envType=problem-list-v2&envId=2cktkvj)**

题目描述：**给你一个单链表的头节点 head ，请你判断该链表是否为回文链表。如果是，返回 true ；否则，返回 false 。**

**第二步，leccode官网思路**<br>

**方法一：将值复制到数组中后用双指针法**<br>
&emsp;&emsp;一共两个步骤：
1. 复制链表值到数组列表中。遍历链表将值复制到数组列表中，使用currentNode指向当前节点，每次迭代向数组添加val的值，更新currentNode指向下一个节点的指针，直至currentNode为空。如理解初始化链表代码一样。
1. 使用双指针法判断是否为回文。


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        vals = []
        current_node = head
        while current_node is not None:
            vals.append(current_node.val)
            current_node = current_node.next
        left, right = 0, len(vals) - 1
        while left < right:
            if vals[left] != vals[right]:
                return False
            left = left + 1
            right = right - 1
        return True
```
