## leecode 206 **[反转链表](https://leetcode.cn/problems/reverse-linked-list/?envType=problem-list-v2&envId=2cktkvj)**

题目描述：**给你单链表的头节点 head ，请你反转链表，并返回反转后的链表。**<br>

*&emsp;&emsp;leecode官方给题目给的示例都不太完美，里面的图没有终止节点，给的例子是个列表，很容易让新手一头雾水，其实这个问题分两步，第一步：先结合python代码搞懂怎么初始化链表，包含数值内容和指针；第二步：围绕链表内部每个指针操作即可。<br>&emsp;&emsp;可以这样具象理解为，一个排列整齐的队伍，每个人拿着手枪顶着前面一个人的后脑勺，但现在需要完成全体向后转之后，再把枪口顶着你向后转之后前面那个人的后脑勺。就像下面1指着2的后脑勺，2指着3的后脑勺，5的枪没卵用。再根据题目理解，给头节点也就是1，反转链表可以想象为从1开始先向后转，这时候1的枪就没卵用了，让2的枪口指向1，这就完成了一次反转，直至5的没卵用的枪最终指向了4。*<br>

**第一步，先理解链表的初始化。**

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
# 排列队伍
node1,node2,node3,node4,node5 = ListNode(1),ListNode(2),ListNode(3),ListNode(4),ListNode(5)
# 定义枪口指向
node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

# cur 就是current 当前节点
head = node1
cur = head
while cur:
    print(cur.val)
    cur = cur.next
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        cur, pre = head, None
        while cur:
            tmp = cur.next  # 暂存后继节点
            cur.next = pre  # 扩充反转链表
            pre = cur # 更新pre，跟上进度
            cur = tmp # 更新cur， 跟上进度
        return pre
```

**复杂度分析**<br>
时间复杂度$O(N)$：遍历链表使用线性大小时间。<br>
空间复杂度$O(1)$：变量 pre 和 cur 使用常数大小额外空间。<br>

**方法二：递归** 考虑使用递归法遍历链表，当越过尾节点后终止递归，在回溯时修改各节点的 next 引用指向。<br>
*&emsp;&emsp;递归的思想确实很抽象，只能说明理解其中的重点，下面代码中res变量是递归到最深层，从pre变量传递过来的值，然后上一层递归的调用res=rev_fun(cur.next,cur)将pre变量的值赋给res，此后回溯期间res变量一直未变，res即为反转链表的头节点。实在不懂，可以向大模型刨根问底。*

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        def rev_fun(cur, pre):
            if not cur:
                return pre
            res = rev_fun(cur.next, cur)
            cur.next = pre
            return res
        return rev_fun(head, None)
```

**复杂度分析**<br>
时间复杂度$O(N)$：遍历链表使用线性大小时间。<br>
空间复杂度$O(N)$：遍历链表的递归深度达到$N$，系统使用$O(N)$大小额外空间。链表长度为$N$，递归函数 recur(cur, pre) 每处理一个节点就会再调用一次，因此会产生大约 
$N+1$ 层函数调用，每一层调用都需要在栈上保存一些信息（局部变量、返回地址等），所以递归栈占用空间随$N$线性增长。<br>
